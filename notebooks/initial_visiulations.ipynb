# Initial explorations of visulations of MNIST

Lets just recreate the manifold visulations to begin with.

```python
from sklearn.datasets import load_digits

digits = load_digits(n_class=3)
```

```python
# View some digits
import matplotlib.pyplot as plt

plt.axis("off")
for i in range(3):
    plt.subplot(1, 3, i + 1)
    plt.imshow(digits.data[i].reshape(8, 8), cmap="gray")
    plt.axis("off")
```

```python
# Visualise the manifold
from sklearn.manifold import Isomap

iso = Isomap(n_components=3)
projection = iso.fit_transform(digits.data)
```

```python
import matplotlib
import numpy as np

mask = digits.target < 3
plt.scatter(
    projection[mask, 0],
    projection[mask, 1],
    lw=0.1,
    c=digits.target[mask],
    cmap=matplotlib.colormaps.get_cmap("tab10").resampled(3),
)
plt.colorbar(ticks=range(3), label="digit value")
plt.clim(-0.5, 2.5)
plt.axis("off")
```
